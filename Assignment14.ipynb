```python
"""
3. What is a box tuple, and how does it work?

Box tuple represents a rectangular region in an image, represented by 4 integer-tuple(left, upper, right, lower)co-ordinates
used in cropping,pasting and resizing of an image
"""

image=Image.open("Passport size photo.jpg")

box_tuple=(100,200,300,400)

im=image.crop(box_tuple)

im.save("cropped.png")
```

```python
"""
7. What module contains Pillow’s shape-drawing code?
ImageDraw module in PIL module
"""
from IPython.display import display
from PIL import Image, ImageDraw

image= Image.new(mode='RGB', size=(500,500), color='Black')

draw=ImageDraw.Draw(image)

draw.line(xy=[(50, 50), (350, 250)], fill='Red', width=5)
draw.ellipse(xy=[(100, 100), (300, 300)], fill='White',outline='Green',width=3)
draw.polygon(xy=[(50, 50), (100, 100), (200, 150), (250, 100), (300, 50)],outline='blue', width=1)

display(image)
```
